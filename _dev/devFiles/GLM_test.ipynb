```python
import patsy
# produce design matrices from R-style formula
formula = 'ofp ~ hosp + health + numchron + gender + school + privins'
```

```python
y, X = patsy.dmatrices(formula, df, return_type='dataframe')
```

```python
poisson_mod = sm.GLM(endog = y, exog = X, family = sm.families.Poisson())
```

```python
poisson_results = poisson_mod.fit()
```

```python
print(poisson_results.summary())
```

The difference is due to following reason:  
"The Negative Binomial distribution has two parameter, besides the mean parameters there is also a dispersion parameter. GLM cannot estimate it in the model, because it implements only a one-parameter linear exponential family. The user needs to provide this parameter, and it's possible to get the MLE in an extra loop outside of GLM.
discrete.NegativeBinomial estimates the dispersion parameter jointly with the mean parameters by MLE. In econometrics two parameterization for the dispersion function and has exponential distribution as special case. Only one of the parameterization corresponds to the GLM parameterization."  
This is discussed here:
https://github.com/statsmodels/statsmodels/issues/3339

```python
negBinom_mod2 = sm.NegativeBinomial(endog = y, exog = X)
negBinom2_results = negBinom_mod2.fit()
print(negBinom2_results.summary())
```

Ideally we'd like to use similar functions for NB and Poisson. Comparing results with sm.Poisson() for Poisson regression:

```python
negBinom2_results.alpha
```

```python
poisson2_mod = sm.Poisson(endog = y, exog = X)
poisson2_results = poisson2_mod.fit()
print(poisson2_results.summary())
```

The results are similar.  
##### Final decision: Choose sm.NegetiveBinomial() and sm.Poisson() for starting values.
